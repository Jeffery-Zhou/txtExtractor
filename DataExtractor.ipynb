```python
import os
import openpyxl
from openpyxl import load_workbook
```

```python
def getColumns():
    f = open("./FundamentalData/TXT2002/AALI0203.TXT","r")
    lines = f.readlines()
    column_lst = []
    for line in lines:
        if ":" in line:
            tmp_lst = line.strip().split(":")
            column_lst.append(tmp_lst[0].strip())
    return column_lst
column_lst = getColumns()
```

```python
def generateExcel(sheet_list):
    wb = openpyxl.Workbook()
    wb.create_sheet(index=0, title="ExtractedData") #index sheet postion
    ws = wb["ExtractedData"]
    ws.cell(1, 1, value = 'Index')
    for i in range(1, len(sheet_list)+1): 
        ws.cell(1,i+1, value = sheet_list[i-1])
    wb.save(filename='./ShuqiWangOwnMeMeal.xlsx')
generateExcel(column_lst)    
```

```python
path_folder_lst = ["./FundamentalData/TXT2002", "./FundamentalData/TXT2003"]
def getPathList(lst):
    path_lst = []
    for i in lst:
        files= os.listdir(i)
        for file in files:
            if not os.path.isdir(file):  
                path_lst.append(os.path.join(i, file))
    return path_lst

path_lst = getPathList(path_folder_lst)
```

```python
len(path_lst)
```

```python
def writeExcel(path):
    wb = openpyxl.load_workbook("./ShuqiWangOwnMeMeal.xlsx")
    ws = wb["ExtractedData"]
    
    f = open(path,"r")
    lines = f.readlines()
    data_lst = []
    index_name = lines[3].strip()
#     print(index_name)
    for line in lines:
        if ":" in line:
            tmp_lst = line.strip().split(":")
            data_lst.append(tmp_lst[1].strip())        
    ws.append([index_name] + data_lst)
    wb.save("./ShuqiWangOwnMeMeal.xlsx")
```

```python
for i in path_lst:
    writeExcel(i)
```
